```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import csv
```

```python
def create_graph(path):
    d={}
    index=1
    with open(path,"r") as csvFile:
        reader=csv.reader(csvFile)
        for row in reader:
            d[row[0]]=index
            index+=1

    df=pd.Series(d)
    df.plot()
    plt.xlabel("seconds")
    plt.ylabel("# of requests")
    plt.title("scale time")
```

```python
create_graph("log.csv")
```
